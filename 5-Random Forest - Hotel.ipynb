# 피처 엔지니어링

```python
ht_df2['people'] = ht_df2['adults'] + ht_df2['children'] + ht_df2['babies']
```

```python
ht_df2.head(3)
```

```python
ht_df3 = ht_df2[ht_df2['people'] != 0]
```

```python
ht_df3['total_nights'] = ht_df3['stays_in_week_nights'] + ht_df3['stays_in_weekend_nights']
```

```python
ht_df3[ht_df3['total_nights'] == 0]
```

```python
ht_df3['season'] = ht_df3['arrival_date_month'].apply(lambda x: 'spring' if x in ['March', 'April', 'May'] else 'summer' if x in ['June', 'July', 'August'] else 'fall' if x in ['September', 'October', 'November'] else 'winter')
```

```python
ht_df3['expected_room_type'] = (ht_df3['reserved_room_type'] == ht_df3['assigned_room_type']).astype('int') #  .()타입으로 변경
```

```python
ht_df3['cancel_rate'] = ht_df3['previous_cancellations'] / ht_df3['previous_cancellations'] + ht_df3['previous_bookings_not_canceled']
```

```python
ht_df3[~ht_df3['cancel_rate'].isna()]
```

```python
ht_df3['cancel_rate'] = ht_df3['cancel_rate'].fillna(-99)
```

# 더미 변수 변환

```python
ht_df3['adr'].dtype
```

```python
obj_list = []

for i in ht_df3.columns:
  if ht_df3[i].dtype == 'O':
    obj_list.append(i)
```

```python
obj_list
```

```python
for i in obj_list:
  print(i, ht_df3[i].nunique())
```

```python
ht_df3 = ht_df3.drop('country', axis= 1)
```

```python
obj_list.remove('country')
```

```python
ht_df3 = pd.get_dummies(ht_df3, columns= obj_list)
```

```python
ht_df3['meal'].value_counts()
```

# 모델링

```python
from numpy.random.mtrand import RandomState
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ht_df3.drop('is_canceled', axis= 1), ht_df3['is_canceled'], test_size= 0.4, random_state= 100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
pred1 = rf.predict(X_test)
```

```python
proba = rf.predict_proba(X_test)
```

```python
proba1 = proba[:, 1] # 1이 될 확률
```

```python
pred1
```

```python
proba1
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
accuracy_score(y_test, pred1)
```

```python
confusion_matrix(y_test, pred1)
```

```python
print(classification_report(y_test, pred1))
```

```python
roc_auc_score(y_test, proba1) # 0.5~ 1.0 까지 1에 가까울수록 좋은 예측
```
