# Train/Test set 나누기

```python
X = rent_df.drop('Rent', axis=1)
y = rent_df['Rent']
```

```python
X
```

```python
pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=80) # (독립변수, 종속변수, 테스트셋 비율, 랜덤모듈 번호)
```

```python
X_train
```

# 모델링

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression() # 모델 변수지정
```

```python
lr.fit(X_train, y_train) # 학습 시키기
```

```python
pred = lr.predict(X_test) # 실제값 테스트
```

```python
y_test # 실제값과 비교
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, pred) # 차이의 절대값 평균 MAE
```

```python
mean_squared_error(y_test, pred, squared= False) # 차이의 제곱 평균 , 루트 [MSE, RMSE - 예측이 중요할때]
```

# 아웃라이러 제거

# Log 활용하기

```python
a = [1,2,3,4,5]
b = [1,10,100,1000,10000]
```

```python
sns.lineplot(x = a, y = b) # 직선이 아니면 예측이 어려움
```

```python
b_log = np.log(b) # 로그로 데이터 변형
```

```python
sns.lineplot(x = a, y = b_log) # 직선 만듬
```

```python
b_log
```

```python
np.exp(b_log) # 지수를 적용 - 원값으로 복원
```

```python
y_train_log = np.log(y_train) # 로그
```

```python
lr.fit(X_train, y_train_log) # 모델링
```

```python
nnewpred = lr.predict(X_test) # 실제값
```

```python
pred_exp = np.exp(nnewpred) # 복원
```

```python
mean_squared_error(y_test, pred_exp, squared= False) # RMSE 비교
```
